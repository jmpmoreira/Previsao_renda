```python
X_train_base, X_test_base = train_test_split(renda, train_size=0.8, random_state=100)
```

### Rodando o modelo


<span style="color:red">ESCREVER AQUI</span>
Neste pedaço, iremos avaliar diferentes modelos de regressão e ver qual se adequa melhor ao modelo

```python
#avaliação das respostas dentro das variáveis 
print(renda['tipo_renda'].value_counts())
print(renda['educacao'].value_counts())
print(renda['estado_civil'].value_counts())
print(renda['tipo_residencia'].value_counts())
```

```python
X = patsy.dmatrices('''renda ~ C(sexo) 
                    + C(posse_de_veiculo) 
                    + C(posse_de_imovel)
                    + qtd_filhos 
                    + C(tipo_renda) 
                    + C(educacao, Treatment(2)) 
                    + C(estado_civil) 
                    + C(tipo_residencia, Treatment(1)) 
                    + idade
                    + tempo_emprego
                    + qt_pessoas_residencia 
                    + 1''', X_train_base)
```

```python
modelo = smf.ols(X, data = X_train_base).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 0.8
                         , alpha = 0.1)
modelo.summary()
```

```python
#calculo da  correlação e o coeficiente de determinação (R²)
X_test_base['pred_renda'] = modelo.predict(X_test_base)
r = X_test_base[['pred_renda', 'renda']].corr().iloc[0,1]
r**2
```

```python
#alterando a variável resposta do log (objetivo de tentar melhorar o modelo)
X = patsy.dmatrices('''np.log(renda) ~ C(sexo) 
                    + C(posse_de_veiculo) 
                    + C(posse_de_imovel)
                    + qtd_filhos 
                    + C(tipo_renda) 
                    + C(educacao, Treatment(2)) 
                    + C(estado_civil) 
                    + C(tipo_residencia, Treatment(1)) 
                    + idade/
                    + tempo_emprego
                    + qt_pessoas_residencia 
                    + 1''', X_train_base)
```

```python
modelo = smf.ols(X, data = X_train_base).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 0.8
                         , alpha = 0.1)
modelo.summary()
```

```python
X_test_base['log_renda'] = np.log(X_test_base['renda'])
X_test_base['pred_renda'] = modelo.predict(X_test_base)
r = X_test_base[['pred_renda', 'log_renda']].corr().iloc[0,1]
r**2
```

Podemos observar uma diminuição de overfitting entre os modelos, ao utilizar o log em relação a renda

Iniciando a preparação para a construção de uma árvore de regressão

```python
#separação das variáveis X_renda e y_renda
X_renda = renda_dummies.drop(columns = ['renda']).copy()
y_renda = np.log(renda_dummies['renda'])
```

```python
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(X_renda, y_renda, train_size=0.8, random_state=100)
```

```python
reg_tree = DecisionTreeRegressor(max_depth=2, min_samples_leaf=10)

reg_tree.fit(X_train_tree, y_train_tree)

dot_data = tree.export_graphviz(reg_tree, out_file=None, 
                                feature_names=X_renda.columns,
                                filled=True)

graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
X = patsy.dmatrices('''np.log(renda) ~ C(sexo)
                    + C(posse_de_veiculo)
                    + C(posse_de_imovel)
                    + qtd_filhos 
                    + idade
                    + tempo_emprego
                    + 1''', X_train_base)

modelo = smf.ols(X, data = X_train_base).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 0.8
                         , alpha = 0.1)
modelo.summary()
```
